```python
import pandas as pd 
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.tail()
```

```python
# koşullu seçimde loop kullanımı:

for i in df["age"]:
    if i > 40:
        print("büyük")
    else:
        print("değil")
```

```python
# Eğer ben burada pandası kullanarak yazmak istersem ilk önce şartı kontrol ediyor olacağım. Sonrasında da koşulu kontrol ediyor olacağım. 


df[df["age"]>50]
```

```python
# burada 50den büyük olanların sayısına ulaşmak istersem nasıl ulaşırım, sorusunun cevabı:

df[df["age"]>50]["age"].count()
```

```python
# Eğer bir koşul ve iki tane sutün seçmek istersem, bunu nasıl yaparım?
# example: Yaşı 50den büyük olan yolcuları ve classlarını görmek istiyorum. 
# loc: label based selection
df.loc[df["age"]>50, ["age", "class"]].head()
```

```python
# Example: Yaşı 50 den büyük olan erkekleri seçmek istiyorum. Bunujn için bütün şartları parantez içerisinde belirtmek zorundasın... 

df.loc[(df["age"]>50) & (df["sex"] == "male") & (df["embark_town"] == "Cherbourg"), ["age", "sex", "class"]].head()

```
