```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
# pivot table ilgilendiğimiz özet istastistliği görmemizi sağlar. 
# şimdi pivot_table ön tanımında mean bulunur, bu şunu ifade eder, senin ilk girdiğin değerin ortalaması alınacaktır.

df.pivot_table("survived", ["sex", "age"], "embarked")
```

```python
# eğer ben pivot tablosunun ön değişlkenin mean olmasını istemiyor burada farklı bir yapıda olmasını istiyorsam, yapmam gereken düzeltme şu şekilde olacaktır.

# Example: burada ben ortalamasını değil de standart sapmasını yapmak istiyorsam:

df.pivot_table("survived", "sex", "embarked", aggfunc="std")
```

```python
# Question: 'titanic' veri seti içerisindeki değerleri kıyasladığım zaman, yaş değişkenin nasıl bir rol oynadığını görmek istiyorum.

df.pivot_table("survived", "age", ["sex", "class"])
```

```python
# cut ve qcut fonksiyonları elimizdeki sayısal değişkenleri kategorik değişkenlere çevirmek için kullanılan en yaygın fonksiyonlardır. 

# eğer ben elimdeki veriyi tanıyorsam (örneğin yaş verileri için 0-10 arası çocuk, 11-18 arası ergen, 18-45 orta, 45-65 yaşlı, 65+ çok yaşlı olarak tanımlamak için cut)

# ama eğer elimdeki verileri tanımıyorsam, çeyreklik halde bölmesini istersem qcut fonbksiyonu kullanılır.
```

```python
df["new_age"] = pd.cut(df["age"], [0,10,18,25,40,65,90])
df.head()
```

```python
df.pivot_table("survived", "sex", ["new_age", "class"])
```
