```python
# Eğer bir değişken verilerini silmek istersem ne yapmalıyım?
# Bunun için drop yöntemini kullanabilirim. drop yönteminde kullanmak istediğim verileri ise şu şekilde yapmalıyım. (...., axis=1), buraadaki axis bir kolonların silinmesi gerektiği anlamına gelecektir. 
```

```python
df["age2"] = df["age"] ** 2
```

```python
df.head()
```

```python
df.drop("age2", axis=1)
```

```python
df["age2"] = df["age"] ** 2 
df["age3"] = df["age"] ** (1/2)
df["age4"] = df["age"] ** 3 
df["age5"] = df["age"] ** (1/3)   
```

```python
df.head()
```

```python
df.loc[:, df.columns.str.contains("age")]

# bu yöntemle içerisinde "age" ifadesi geçen her şeyi seçebildik. 
```

```python
# burada bir ifade dışındaki değişkenleri seçebilmek için neler yapabiliriz,  bunları göz önüne aldığımızda pythonda değildir ifadesi olan 'tilta' işaretini (~) (alt gr + ü ) işaretine basarak yapabiliriz.

# Example: Eğer içerisinde age bulunduranların dışındaki değişkenleri seçmek istiyorsak:

df.loc[:, ~df.columns.str.contains("age")]
```
