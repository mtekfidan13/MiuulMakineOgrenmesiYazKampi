```python
import pandas as pd
import seaborn as sns
```

```python
# loc ve iloc işlemleri dataframelerde seçimler için kullanılacak olan özel yapılardır.
# iloc index bilgisi vererek seçim yapmamızı sağlar. 
# loc ise indexteki labellara göre seçim yapmamızı sağlar. 

df = sns.load_dataset("titanic")
```

```python
df.iloc[0:123] ## başlangıç ve bitiş indexlerini yazıyorsun, yani index'e göre sıralıyorsun 
```

```python
# örndeğin tam konumdaki elemanı merak ediyorsun, orada [,] şeklinde yapabilirsin

df.iloc[0,1]
```

```python
# loc : label based selection

df.loc[0:3] # burada virgülden sonra yazdığın  değerde dahil edilior. Buna dikkat et.

```

```python
# iloc: integer label based selection, yani burada tam index gireceksin
# Example: 

df.iloc[0:3,0:4]
```
