```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
```

```python
## Plot özelliği veriyi görselleştirmek amacıyla kullandığımız görsellerden bir tanesidir. 

x = np.array([1,8])
y = np.array([0,150])

plt.plot(y,x)
```

```python
plt.plot(x,y,'o')
```

```python
### marker özellikleri 
y = np.array([13,28,11,100])
plt.plot(y, marker='o')
```

```python
plt.plot(y, marker='*')
```

```python
## Line özelliği 
y = np.array([13,28,11,100])
plt.plot(y, linestyle="dashdot", color="r")
```

```python
## Multiple Lines

x = np.array([19,25,74,100])
y = np.array([91,52,69,10])

plt.plot(x, color="r")
plt.plot(y, color="b")
```
